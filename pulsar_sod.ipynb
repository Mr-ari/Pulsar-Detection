```python
from pyod.models.sod import SOD
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Documents/Pulsar-Detection/HTRU2/HTRU_2.csv')
```

```python
df.head()
```

```python
pulsar = pd.DataFrame(df.drop('class_labels', 1))
```

```python
pulsar.head()
```

```python
clf = SOD(contamination=0.1)
```

```python
clf.fit(pulsar)
```

```python
df['scores'] = clf.decision_scores_
```

```python
df['labels'] = clf.labels_
```

```python
plt.hist(df['scores'])
```
