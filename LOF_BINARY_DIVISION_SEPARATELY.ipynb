```python
from pyod.models.lof import LOF
from sklearn.neighbors import LocalOutlierFactor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pulsar = pd.read_csv('Documents/Pulsar-Detection/HTRU2/HTRU_2.csv')
```

```python
pul1 = pulsar.copy()
pul2 = pulsar.copy()
pul1 = pul1.drop(['Mean_of_the_DM-SNR_curve', 'Standard_deviation_of_the_DM-SNR_curve', 'Excess_kurtosis_of_the_DM-SNR_curve', 'Skewness_of_the_DM-SNR_curve', 'class_labels'], axis=1)
pul2 = pul2.drop(['Mean_of_the_integrated_profile', 'Standard_deviation_of_the_integrated_profile', 'Excess_kurtosis_of_the_integrated_profile', 'Skewness_of_the_integrated_profile', 'class_labels'], axis=1)
```

```python
clf = LocalOutlierFactor(n_neighbors=5,contamination=0.1)

clf.fit(pul1)
scr1 = clf.negative_outlier_factor_

clf.fit(pul2)
scr2 = clf.negative_outlier_factor_
```

```python
pul1['class_labels'] = pulsar['class_labels']
pul1['score'] = scr1
pul2['class_labels'] = pulsar['class_labels']
pul2['score'] = scr2
```

```python
sorted_val1 = pul1.sort_values(by = 'score' , ascending = True)
sorted_val2 = pul2.sort_values(by = 'score' , ascending = True)
```
